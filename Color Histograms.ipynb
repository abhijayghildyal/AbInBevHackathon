```python
beerImages = os.listdir('beerImages')
print beerImages
```

```python
beerHistograms = {}
```

```python
for beerImage in beerImages:
    print ("\nBeer: "+beerImage)
    img = cv2.imread('beerImages/'+beerImage)
    beerHistograms[beerImage] = cv2.calcHist([img],[0],None,[256],[0,256])
    #plt.imshow(img)
    plt.hist(img.ravel(),256,[0,256]);
    plt.show()
    print("-----------------------------------------------------------------\n")
```

```python
beerHistograms.keys()
```

```python
distanceMatrix = np.zeros(shape=(20,20))
```

```python
for i in xrange(0,len(beerImages)):
    for j in xrange(0,len(beerImages)):
        distanceMatrix[i][j] = (np.linalg.norm(beerHistograms[beerImages[i]]-beerHistograms[beerImages[j]]))
```

```python
pd.DataFrame( distanceMatrix, index=beerHistograms.keys(), columns=beerHistograms.keys())
```
